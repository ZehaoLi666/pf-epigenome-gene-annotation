```python
from deeptools.computeMatrix import main as computeMatrix
from deeptools.plotHeatmap   import main as plotHeatmap
from deeptools.plotProfile   import main as plotProfile
import os

regions1 = os.path.expanduser("~/lab/LncRNA_chip_prediction/Final/window1000_bin100/New_genome_features/Deeptools/predicted_genes_TSS.filtered_by_geneid.robust.bed")
regions2 = os.path.expanduser("~/lab/LncRNA_chip_prediction/Final/window1000_bin100/New_genome_features/Deeptools/TSS_5UTRgt500.fromGFF.bed")

# 1) computeMatrix
computeMatrix([
  "reference-point","--referencePoint","TSS",
  "-S","/rhome/zli529/lab/SRA_toolkit/RNAseq_7stages/processed_data/LT/RNA.bw",
  "-R", regions1, regions2,
  "-a","1000","-b","1000","-bs","10",
  "--skipZeros","--missingDataAsZero",
  "--numberOfProcessors","8",
  "-o","RNA_matrix_TSS.gz",
  "--outFileNameMatrix","RNA_matrix_TSS.tab",
  "--outFileSortedRegions","RNA_regions_TSS.bed"
])

# 2) Heatmap (per region group)
plotHeatmap([
  "-m","RNA_matrix_TSS.gz",
  "--colorMap","viridis","--perGroup",
  "--plotTitle","RNA ±500 bp around TSS",
  "-out","RNA_heatmap_TSS.svg"
])

# 3) Profile (per region group)
plotProfile([
  "-m","RNA_matrix_TSS.gz",
  "--perGroup","--plotType","lines",
  "--samplesLabel","RNA",
  "-out","RNA_profile_TSS.svg"
])

from IPython.display import SVG, display
display(SVG("RNA_heatmap_TSS.svg"))
display(SVG("RNA_profile_TSS.svg"))
```
